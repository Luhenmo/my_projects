```python
from data_base import GLOBAL_DATA_BASE
from pathlib import Path
from classes import Portfolio
import numpy as np
from copy import deepcopy

current_path = Path.cwd()
GLOBAL_DATA_BASE = GLOBAL_DATA_BASE
```

```python
portfolio = Portfolio(GLOBAL_DATA_BASE,"NC")
portfolio_global = deepcopy(portfolio)
portfolio_global.data_base.columns = ["Dono","Data","Ativo","C/V","Preço","Qnt","Moeda","Classe"]
mask1 = portfolio_global.data_base["Classe"] == "Tesouro"
mask2 = portfolio_global.data_base[mask1]["Ativo"] == "LFTS11"
display(portfolio_global.data_base[mask1])
# Create a combined mask
mask = mask1 & (~mask2)

# Update 'Qnt'
portfolio_global.data_base.loc[mask, "Qnt"] = (
    portfolio_global.data_base.loc[mask, "Qnt"].apply(lambda value: 100 * value)
)

# Update 'Preço'
portfolio_global.data_base.loc[mask, "Preço"] = (
    portfolio_global.data_base.loc[mask, "Preço"].apply(lambda value: np.round(value / 100, 2))
)
portfolio_global.data_base["Qnt"] = portfolio_global.data_base["Qnt"].apply(int)
display(portfolio_global.data_base[mask1])
dict_tipo = {True:"C",False:"V"}
portfolio_global.data_base["C/V"] = portfolio_global.data_base["C/V"].apply(lambda value: dict_tipo[value])

portfolio_global.data_base.drop("Classe",axis=1,inplace=True)

portfolio_global.data_base.to_csv(current_path/"historico_transacoes.csv", index=False)
```
